```python
f_x.subs(x,-4.79128784747792)
```

```python
K = sp.Matrix([[1,1,1],[1,2,2],[2,1,3]])
```

```python
x1,x2,x3 = sp.symbols(['x1','x2','x3'])
```

```python
K
```

```python
q = sp.Matrix([x1,x2,x3])
```

```python
F = sp.Matrix([6,9,11])
```

```python
# K*q - F = {0} ==> K*q = F
```

```python
sistema = K*q - F
sistema
```

```python
sp.solve(sistema,(x1,x2,x3))
```
