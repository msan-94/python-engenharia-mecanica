## 3. Métodos importantes

```python
#.upper() transforma a string correspondente em letras maiúsculas

nome = "Matheus"
print(nome.upper())
```

```python
#.lower() transforma a string correspondente em letras minúsculas

nome = "Matheus"
print(nome.lower())
```

```python
#.strip() remove espaço"
texto = " oi "
print(texto.strip())
```

```python
#.replace() troca o texto original pelo segundo termo proposto dentro dos parêntesis.
texto = "motor quebrado"
print(texto.replace("quebrado", "ok"))
```

## 4. Comparação de strings (MUITO IMPORTANTE)

```python
resposta = input("Deseja parar? ")

if resposta == "S":
    print("Parando")

# ...mas o operador pode digitar S, s ou s (com espaço)
```

```python
# Então utilizamos o seguinte artifício:

resposta = input("Deseja parar? ").strip().upper()
if resposta == "S":
    print("Parando")
```

# 6. Mini desafio - Melhorar o sistema anterior
### Aceitar s, S, sim, etc...
### Limpar entrada



```python
def mediafreq(freq):
    soma = 0
    for i in freq:
        soma += i
    return soma / len(freq)


while True:
    freq = []
    
    for i in range(5):
        freq.append(float(input("Digite o valor da frequencia (Hz): ")))
    
    media = mediafreq(freq)
    
    if media > 250:
        print("CRÍTICO")
    elif media > 150:
        print("ALERTA")
    else:
        print("SISTEMA OPERANTE")
    
    STOP = input("Deseja parar? (S/N): ").strip().upper()
    
    if STOP == "S" or STOP == "SIM":
        break
    else:
        print("Continuando")
```
