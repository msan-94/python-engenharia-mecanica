## 3. Engenharia básica (velocidade)

```python
dist = int(input('Digite a distância em metros: .'))
temp = float(input('Digite o tempo em segundos: .'))

vel = dist/temp

print('A velocidade é de', vel, 'm/s')
```

## 4. Engenharia básica (força)

```python
massa = int(input('Digite a massa em Kg: '))
acel = float(input('Digite a aceleração em m/s2: '))

forca = massa*acel

print('A força é de', forca, 'kgf.')
```
