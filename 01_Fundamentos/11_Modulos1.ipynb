# 5. Mini desafio
## Simular variação no sensor
### Adicionar ruído aleatório na medição

```python
def mediafreq(freq):
        soma = 0
        for i in freq:
            soma += i
        return soma / len(freq)

while True:
    freq = []
    
    for valor in range(5):
        
        while True:
            try:
                valor = float(input("Digite o valor da frequencia (Hz): "))
                ruido = random.uniform(-0.5, 0.5)
                valor_real = valor + ruido
                freq.append(valor_real)
                break
            except ValueError:
                print("Digite apenas números!")
          
    media = mediafreq(freq)
    
    print(freq)
    print(media)
    
    if media > 250:
        print("CRÍTICO")
    elif media > 150:
        print("ALERTA")
    else:
        print("SISTEMA OPERANTE")
    
    STOP = input("Deseja parar? (S/N): ").strip().upper()
    
    if STOP == "S" or STOP == "SIM":
        break
    else:
        print("Continuando")
```
