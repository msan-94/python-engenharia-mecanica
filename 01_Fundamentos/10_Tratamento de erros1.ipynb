## 1. O problema

```python
valor = float(input("Digite: "))
```

```python
# Se o usuário digitar abc ----> ValueError
```

## 2. Solução

```python
try:     # tenta executar
    valor = float(input("Digite: "))
except:    # executa se der erro
    print("Erro de entrada")
```

## 3. Forma mais específica

```python
try:     # tenta executar
    valor = float(input("Digite: "))
except:    # executa se der erro
    print("Digite apenas números")
```

## 4. Loop com validação 

```python
while True:
    try:
        valor = float(input("Digite: "))
        break
    except ValueError:
        print("Entrada inválida")
```

# Exercícios

## 1.Básico

```python
try:
    idade = int(input("Digite a sua idade: "))
    print(idade)
except ValueError:
    print("Apenas números")
```

## 2. Loop seguro

```python
while True:
    try: 
        frq = float(input("Digite a frequência: "))
        print(frq)
        break
    except ValueError:
        print("Apenas números")
```

## 3. Sensor seguro

```python
valores = []

for i in range(3):
    while True:
        try:
            v = float(input("Sensor: "))
            valores.append(v)
            break
        except ValueError:
            print("Erro no sensor")
print(valores)
```

## 4. Divisão segura

```python
try:
    a = float(input("Digite um número: "))
    b = float(input("Digite um número: "))
    
    print(a/b)
except ZeroDivisionError:
    print("Divisão por zero!")
```

# 5. Mini desafio
## Melhorar sistema com proteção contra erro
### A) Evitar erro de input
### B) Garantir dados válidos

```python
def mediafreq(freq):
        soma = 0
        for i in freq:
            soma += i
        return soma / len(freq)

while True:
    freq = []
    
    for valor in range(5):
        
        while True:
            try:
                valor = float(input("Digite o valor da frequencia (Hz): "))
                freq.append(valor)
                break
            except ValueError:
                print("Digite apenas números!")
          
    media = mediafreq(freq)
    
    if media > 250:
        print("CRÍTICO")
    elif media > 150:
        print("ALERTA")
    else:
        print("SISTEMA OPERANTE")
    
    STOP = input("Deseja parar? (S/N): ").strip().upper()
    
    if STOP == "S" or STOP == "SIM":
        break
    else:
        print("Continuando")
```
