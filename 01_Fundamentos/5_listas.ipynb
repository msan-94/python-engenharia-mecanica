## 1. O que é uma lista?

```python
temp = [50, 60, 70, 80, 90] #uma lista guarda vários valores em uma única variável.
```

## 2. Acessando valores

```python
print(temp[0]) #o primeiro valor é identificado como "0" o segundo como "1", o terceiro como "2", e assim por diante.
print(temp[1]) #para acessar: função + variável + índice (0,1,2,3...).
```

## 3. Percorrendo uma lista

```python
for t in temp: #traduzindo: para cada valor "t" na variável, que no caso é uma lista "temp", imprima os valores "t".
    print(t)
```

## Exercício 2 (coleta de dados)

```python
temp2 = []

for i in range(5): #para cada valor "i" com o range de 5.
    temp3 = float(input('temp4: ')) #crie uma variável "temp3" que contenha outra variável que 
    #preencherá os 5 valores solicitado.
    temp2.append(temp3) #construa uma lista com os valores de "temp3"

print(temp2) #imprima a lista
```

## Exercício 3 (média com listas)

```python
temp5 = [50, 60, 70, 80, 90] #lista proposta

soma = 0 #variável usada para armazenar a soma das temperaturas

for tt in temp5: #para cada valor "tt" dentro da lista proposta...
    soma += tt #soma tt a variável soma

media = soma / len(temp5) #cálculo da média = soma das temperaturas / pela quantidade de temperaturas - 
#usando a função len() para projetar a quantidade de temperaturas na lista.

print(media) #imprimir temperatura média
```

## Execício 4 (integração com lógica)

```python
temp6 = [50, 60, 70, 80, 90]

soma1 = 0

for tt1 in temp6:
    soma1 += tt1
    
med = soma1 / len(temp6)

if med >= 80:
    print("CRÍTICO")
elif med >= 60:
    print("ALERTA")
else:
    print("NORMAL")
```

## Mini desafio
#### 1. Ler 5 temperatura
#### 2. Armazenar em lista
#### 3. Calcular a média
#### 4. Classificar

```python
#executando apenas 5 mediçoes de temperatura e finalizando o programa.

tempx = []

for i in range(5): 
    tempx.append(float(input("Digite o valor da temperatura: ")))

soma = 0

for t0 in tempx:
    soma += t0

med0 = soma / len(tempx)

print(med0)

if med0 >= 150:
    print("CRÍTICO")
elif med0 >= 100:
    print("ALERTA")
else:
    print("NORMAL")
```

```python
#executando mediçoes até que o comando STOP seja ativado.

freqs = []

while True:
    
    freqs = []
    
    for i in range(5):
        freqs.append(float(input("Digite o valor da frequencia: ")))

    somaf = 0

    for f in freqs:
        somaf += f
                     
    medf = somaf / len(freqs)
    
    print("A frenquencia média é: ", medf)
                     
    if medf >= 150:
        print("CRÍTICO")
    elif medf >= 100:
        print("ALERTA")
    else:
        print("NORMAL")                     
                     
    stop = input("Deseja parar? (S/N): ")

    if stop == "S":
        print("Parando sistema de mediçao")
        break
    
    else:
        print("continuando")
```
