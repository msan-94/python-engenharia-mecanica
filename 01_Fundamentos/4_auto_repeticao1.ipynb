## 1. Loop for (repetição controlada)

```python
for i in range(5): #range 5 --> vai de 0 até 4
    print(i) #i é a variavel que muda
```

## 2. Loop while (repetição com condição)

```python
contador = 0

while contador < 5:
    print(contador)
    contador += .5
```

## 3. while

```python
numero = 0

while numero < 5:
    print(numero)
    numero +=1
```

```python
while True: #sistema rodando continuamente
    t1 = float(input('Digite o valor da temperatura 1: '))
    t2 = float(input('Digite o valor da temperatura 2: '))
    t3 = float(input('Digite o valor da temperatura 3: ')) #leitura do sensor
    t4 = float(input('Digite o valor da temperatura 4: '))
    t5 = float(input('Digite o valor da temperatura 5: '))
    tmed = (t1+t2+t3+t4+t5)/5
    
#lógica de segurança    
    if tmed >= 80:
        print('CRÍTICO') #mensagem no painel do sistema
    elif tmed >= 60:
        print('ALERTA') #mensagem no painel do sistema
    else:
        print('NORMAL') #mensagem no painel do sistema
        
    sair = input('Deseja parar? (s/n): ')
    if sair == 's':
        break
```
