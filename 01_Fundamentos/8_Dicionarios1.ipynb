# 6. Mini Desafio
### 1. Guardar sensores com nome
### 2. Calcular média
### 3. Classíficar

```python
def mediafreq(freq):
    soma = 0
    for chave, valor in freq.items():
        soma += valor
        if len(freq) == 0: 
            return 0
    return soma / len(freq)


while True:
    
    freq = {}
    
    for i in range(2):
        nome = input("Nome do sensor: ")
        valor = (float(input("Digite o valor da frequencia (kHz): ")))
        freq[nome] = valor
    
    media = mediafreq(freq)
    
    if media > 5:
        print("CRÍTICO")
    elif media > 3:
        print("ALERTA")
    else:
        print("SISTEMA OPERANTE")
    
    STOP = input("Deseja parar? (S/N): ").strip().upper()
    
    if STOP == "S" or STOP == "SIM":
        break
    else:
        print("Continuando")
```
