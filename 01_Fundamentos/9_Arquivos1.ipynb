```python
def mediafreq(freq):
    soma = 0
    for chave, valor in freq.items():
        soma += valor
        if len(freq) == 0: 
            return 0
    return soma / len(freq)


while True:
    
    freq = {}
    
    for i in range(2):
        nome = input("Nome do sensor: ")
        valor = (float(input("Digite o valor da frequencia (kHz): ")))
        freq[nome] = valor
    
    media = mediafreq(freq)
    
    with open("freqs1.txt", "a") as fs:
        fs.write(f"{freq} | media: {media}\n")

    if media > 5:
        print("CRÍTICO")
    elif media > 3:
        print("ALERTA")
    else:
        print("SISTEMA OPERANTE")

    print("Dados registrados")
    
    STOP = input("Deseja parar? (S/N): ").strip().upper()
    
    if STOP == "S" or STOP == "SIM":
        break
    else:
        print("Continuando")
```
