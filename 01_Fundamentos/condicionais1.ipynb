## 1. Estrutura if

```python
temperatura = 80 #criar variável para definir parâmetro.

if temperatura > 70:
    print('Sistema superaquecido')
```

## 2. If + Else

```python
temperatura1 = 50

if temperatura1 > 70: 
    print('Perigo')
else: #O else roda enquanto o if é falso. Para toda temperatura que nao satisfaça o if, 
    #ou seja tudo o que for menor ou igual à condição do if.
    print('Normal')
```

## 3. elif (vários casos)

```python
temperatura2 = 70

if temperatura2 > 80:
    print('Crítico')
elif temperatura2 > 60:
    print('Atenção')
else:
    print('Seguro')
```

## 4. Comparações

```python
# > maior
# < menor
# >= maior ou igual
# <= menor ou igual
# == igual
# != diferente
```

# Exercícios

## 1. Solicitar temperatura

```python
temp1 = int(input('Digite a temperatura da caldeira '))

temp2 = int(input('Digite a temperatura do sistema de refrigeração '))
 
t3 = temp1 + temp2
            
if t3 >= 350:
    print('Perigo!!!!!!!')
elif t3 >= 300:
    print('Atençao')
else:
    print('Sistema seguro')
            
```

## 2. Solicitar pressão

```python
press1 = float(input('Digite a pressão na câmara1 em bar: '))
press2 = float(input('Digite a presão na câmara2 em bar: '))

press3 = press1+press2

if press3 <= 120:
    print('Sistema comprometido')
elif 120 < press3 < 240:
    print('Sistema operacional')
elif press3 > 240:
    print('Alerta de sobrecarga')
    
```

## 3. Validar entrada

```python
vel = float(input('Digite a velocidade em m/s: '))

if vel < 0:
    print('Valor inválido: ')
else:
    print('Valor válido')
```

```python
A = input('Digite um valor para A: ')
B = input('Digite um valor para B: ')

if A < B:
    print('A maior que B')
elif A == B:
    print('A e B são iguais')
else:
    print('B maior que A')
```
