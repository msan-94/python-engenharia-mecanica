# 5. Mini desafio
### Criar tabela
### Calcular média
### Classificar

```python
import pandas as pd

# 3 sensores, 2 medições de cada grandeza física (frequencia e temperatura)

dados = {
    "sens": ["sens_porta", "sens_janela", "sens_piso"],
    "freq": [434, 332, 565],
    "temp": [22, 23, 16]
}
df = pd.DataFrame(dados)

med_freq = df["freq"].mean()

print("Média: ", med_freq)

df["status"] = df["freq"].apply(lambda x: "CRÍTICO" if x >= 350 else "ATENÇÃO" if x > 250 else "NORMAL")

print(df)
```
