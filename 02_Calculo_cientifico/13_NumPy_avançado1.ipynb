## 1. Array multidimensionais

```python
import numpy as np

# Matriz 2D

dados = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

# Acessando elementos | (x = linhas; y = colunas) | começando sempre de 0 (zero). 

print(dados[0, 0]) #10
print(dados[1, 2]) #60
```

## 2. Estatísticas importantes

```python
print(dados.mean()) # media geral
print(dados.max()) # máximo
print(dados.min()) # mínimo
```

## 3. Operações por eixo

```python
# Axis:
# 0 ==== colunas
# 1 ==== linhas

print(dados.mean(axis = 0)) # media por colunas
print(dados.mean(axis = 1)) # media por linhas
```

## 4. Filtros (lógica vetorial)

```python
dados = np.array([10, 50, 100, 200])

print(dados > 80)
```

```python
print(dados[dados > 80])
```

## 5. Operações vetoriais reais

```python
dados = np.array([1, 2, 3])

resultado = dados * 10 + 5

print(resultado)
```

# 5. Mini desafio
### Matriz = sensores x tempo
### Calcular média por sensor
### Detectar falhas

```python
import numpy as np

# 6 sensores, 3 medições

dados = np.array([
    [10, 20, 30], # sensor 1
    [40, 50, 60], # sensor 2
    [70, 80 ,90], # sensor 3
    [100, 110, 120], # sensor 4
    [130, 140, 150], # sensor 5
    [160, 170, 180] # sensor 6
])

med_sens = dados.mean(axis = 1)

print("Média: ", med_sens)

falha = dados[dados > 150]

print("Sensores com falha: ", falha)
```
